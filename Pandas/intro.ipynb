```python
#converting list to DataFrame
data = [1,2,3,4,5]
df = pd.DataFrame(data)
df

```

```python
row,column = df.shape
print("Number of rows:", row)
print("Number of columns:", column)
```

```python
#set column names
df.columns = ['Numbers']
df

```

```python
#or can be done while creating DataFrame
df = pd.DataFrame(data, columns=['Numbers'])
df
```

```python
#names of index means naming the rows
df.index = ['a', 'b', 'c', 'd', 'e']
df
```

```python
#or can be done while creating DataFrame
df = pd.DataFrame(data, columns=['Numbers'], index=['a', 'b', 'c', 'd', 'e'])
df
```

```python
#to create DataFrame with  2 lists
city=['Delhi', 'Mumbai', 'Chennai', 'Kolkata', 'Bangalore']
population=[19.8, 20.4, 7.1, 4.5, 8.4]
data = list(zip(city, population))# zip function combines two lists into a list of tuples
print(data)
df = pd.DataFrame(data, columns=['City', 'Population'])
df
```

```python
city=[['Sylhet','Bangladesh'],['Asam','India'],['Kathmandu','Nepal'],['Thimphu','Bhutan']]
df = pd.DataFrame(city, columns=['City', 'Country'])
df
```
